# ── 3. Copy + Rename Checkpoints ─────────────────────────

```python
from __future__ import annotations
import re
import shutil
import csv
from datetime import datetime

SIZE_RE  = re.compile(r"(yolo11[nsmlx])\.pt")
BATCH_RE = re.compile(r"train_val_test_(\d+)")


def parse_size_and_batch(src: Path) -> tuple[str, str]:
    s = str(src)
    m1 = SIZE_RE.search(s)
    m2 = BATCH_RE.search(s)
    if not m1:
        raise ValueError(f"Cannot parse yolo size from path: {src}")
    if not m2:
        raise ValueError(f"Cannot parse batch from path: {src}")
    return m1.group(1), m2.group(1)


def safe_copy2(src: Path, dst: Path) -> Path:
    dst.parent.mkdir(parents=True, exist_ok=True)
    if not dst.exists():
        shutil.copy2(src, dst)
        return dst
    stem, suffix = dst.stem, dst.suffix
    i = 1
    while True:
        cand = dst.with_name(f"{stem}_v{i}{suffix}")
        if not cand.exists():
            shutil.copy2(src, cand)
            return cand
        i += 1


def build_new_name(size: str, batch: str) -> str:
    return f"{size}_{NAME_MIDDLE}_batch_{batch}_{NAME_SUFFIX}.pt"
```

```python
# ── Step 2：批量 Copy + Rename checkpoint 到目标目录 ───────
# 输入：SRC_PATHS 中的 best.pt 文件
# 输出：DST_DIR 下重命名后的 .pt 文件 + manifest.csv

logger.info("Step 2 开始：批量 copy + rename checkpoint")

DST_DIR.mkdir(parents=True, exist_ok=True)
rows = []
ok = 0
fail = 0

for p in SRC_PATHS:
    src = Path(p)
    try:
        if not src.exists():
            raise FileNotFoundError(f"Missing file: {src}")
        size, batch = parse_size_and_batch(src)
        new_name = build_new_name(size, batch)
        dst = DST_DIR / new_name
        final_dst = safe_copy2(src, dst)
        rows.append({"src": str(src), "dst": str(final_dst), "size": size, "batch": batch,
                     "copied_at": datetime.now().isoformat(timespec="seconds")})
        ok += 1
        print(f"[OK] {src.name}  ->  {final_dst.name}")
    except Exception as e:
        fail += 1
        rows.append({"src": str(src), "dst": "", "size": "", "batch": "",
                     "copied_at": datetime.now().isoformat(timespec="seconds"), "error": str(e)})
        logger.error(f"Copy 失败 {src}: {e}")

manifest_path = DST_DIR / "manifest.csv"
fieldnames = sorted({k for r in rows for k in r.keys()})
try:
    with manifest_path.open("w", newline="", encoding="utf-8") as f:
        w = csv.DictWriter(f, fieldnames=fieldnames)
        w.writeheader()
        w.writerows(rows)
except Exception as e:
    logger.error(f"写入 manifest 失败: {e}")

logger.info(f"Step 2 完成：OK={ok}, FAIL={fail}, DST_DIR={DST_DIR}, manifest={manifest_path}")
```

# ── 4. SAHI 推理 Imports & Helpers ──────────────────────

```python
from __future__ import annotations
from typing import Dict, List, Tuple, Any

import pandas as pd
import fiftyone as fo
import fiftyone.utils.coco as fouc
from fiftyone import ViewField as F

from sahi import AutoDetectionModel
from sahi.predict import get_sliced_prediction
```

```python
def fetch_subsequent_dir(data_root: Path, target_subdir_name: Path) -> Tuple[List[Path], List[str]]:
    data_paths = list(data_root.glob(f"*/{target_subdir_name}"))
    subdir_path_list = [p.parent for p in data_paths]
    subdir_name_list = [d.name for d in subdir_path_list]
    return subdir_path_list, subdir_name_list


def model_tag_from_path(p: str) -> str:
    return Path(p).stem


def ensure_pred_field(dataset: fo.Dataset, pred_field: str) -> None:
    if pred_field not in dataset.get_field_schema():
        dataset.add_sample_field(pred_field, fo.EmbeddedDocumentField,
                                 embedded_doc_type=fo.Detections)


def ensure_dataset_with_gt(subdir_path: Path, subdir_name: str, version: str) -> fo.Dataset:
    ds_name = f"{version}_{subdir_name}"
    if ds_name in fo.list_datasets():
        return fo.load_dataset(ds_name)
    dataset = fo.Dataset.from_images_dir(str(subdir_path / "data"), name=ds_name)
    coco_path = subdir_path / coco_json_name
    if not coco_path.exists():
        raise FileNotFoundError(f"COCO JSON not found: {coco_path}")
    fouc.add_coco_labels(dataset, label_field="ground_truth",
                         labels_or_path=str(coco_path), categories=categories,
                         label_type="detections")
    return dataset


def run_inference_for_model(dataset: fo.Dataset, ckpt_path: str, pred_field: str) -> None:
    detection_model = AutoDetectionModel.from_pretrained(
        model_type="yolov8", model_path=ckpt_path,
        confidence_threshold=0.1, image_size=640, device="cuda",
    )
    ensure_pred_field(dataset, pred_field)
    for sample in dataset.iter_samples(progress=True, autosave=True):
        result = get_sliced_prediction(sample.filepath, detection_model, verbose=0,
                                       slice_height=slice_height, slice_width=slice_width,
                                       **sahi_kwargs)
        sample[pred_field] = fo.Detections(detections=result.to_fiftyone_detections())
```

```python
def evaluate_model_fields(
    dataset: fo.Dataset, ds_name: str, subdir_path: Path,
    ckpt_path: str, model_tag: str, pred_field: str,
    thresholds: List[float], only_matches: bool = False,
) -> List[Dict[str, Any]]:
    rows: List[Dict[str, Any]] = []
    gt_count = dataset.count("ground_truth.detections")
    total_pred_count = dataset.count(f"{pred_field}.detections")

    for thr in thresholds:
        view = dataset.filter_labels(pred_field, F("confidence") > thr, only_matches=only_matches)
        new_model_tag = model_tag.replace("-", "__")
        eval_key = f"eval_{new_model_tag}_conf_{int(thr * 100)}_onlymatches_{only_matches}"
        res = view.evaluate_detections(pred_field, gt_field="ground_truth",
                                       compute_mAP=True, eval_key=eval_key)
        pred_count_thr = view.count(f"{pred_field}.detections")
        metrics = res.metrics()
        rows.append({
            "dataset_name": ds_name, "version": version,
            "model_tag": model_tag, "ckpt_path": ckpt_path, "pred_field": pred_field,
            "confidence_threshold": thr,
            "gt_count": int(gt_count), "pred_count_total": int(total_pred_count),
            "pred_count_at_threshold": int(pred_count_thr),
            "mAP": float(res.mAP()),
            "precision": float(metrics.get("precision", float("nan"))),
            "recall": float(metrics.get("recall", float("nan"))),
            "f1": float(metrics.get("fscore", float("nan"))),
        })
    return rows
```

```python
# ── Step 3：扫描数据子目录，打印路径供 Master 确认 ─────────
# 输入：data_root 下各子目录
# 输出：打印 subdir 列表

logger.info("Step 3 开始：扫描数据子目录")

subdir_path_list, subdir_name_list = fetch_subsequent_dir(data_root, target_subdir_name)
print(f"Found subdirs: {len(subdir_name_list)}")
for n, p in zip(subdir_name_list, subdir_path_list):
    print(f"  {n}  ->  {p}")

logger.info(f"Step 3 完成：共 {len(subdir_name_list)} 个子目录")
```

# ── 6. Build Datasets + 多模型 SAHI 推理 ─────────────────

```python
# ── Step 4：对每个子目录创建 dataset（GT 只导一次），然后跑多模型推理 ──
# 输入：subdir_path_list, ckpt_paths
# 输出：FiftyOne datasets，每个 dataset 含各模型的 pred_field

logger.info("Step 4 开始：Build datasets + 多模型 SAHI 推理")

for subdir_path, subdir_name in zip(subdir_path_list, subdir_name_list):
    try:
        dataset = ensure_dataset_with_gt(subdir_path, subdir_name, version)
    except Exception as e:
        logger.error(f"创建 dataset 失败 {subdir_name}: {e}")
        continue

    ds_name = dataset.name
    logger.info(f"Dataset: {ds_name}, GT count: {dataset.count('ground_truth.detections')}")

    for ckpt_path in ckpt_paths:
        model_tag = model_tag_from_path(ckpt_path)
        pred_field = f"small_slices_{model_tag}"
        logger.info(f"[INFER] model={model_tag}, field={pred_field}")
        try:
            run_inference_for_model(dataset, ckpt_path, pred_field)
            pred_count = dataset.count(f"{pred_field}.detections")
            logger.info(f"[INFER] 完成: {model_tag}, pred_count={pred_count}")
        except Exception as e:
            logger.error(f"推理失败 model={model_tag}, dataset={ds_name}: {e}")

logger.info("Step 4 完成：所有 dataset 推理结束")
```

# ── 7. 评估所有模型 -> DataFrame ─────────────────────────

```python
# ── Step 5：评估多模型多阈值，汇总为 DataFrame ─────────────
# 输入：FiftyOne datasets 中各模型的 pred_field
# 输出：all_rows -> df

logger.info("Step 5 开始：评估所有模型")

all_rows: List[Dict[str, Any]] = []

for subdir_path, subdir_name in zip(subdir_path_list, subdir_name_list):
    ds_name = f"{version}_{subdir_name}"
    try:
        dataset = fo.load_dataset(ds_name)
    except Exception as e:
        logger.error(f"加载 dataset 失败 {ds_name}: {e}")
        continue

    for ckpt_path in ckpt_paths:
        model_tag = model_tag_from_path(ckpt_path)
        pred_field = f"small_slices_{model_tag}"
        if pred_field not in dataset.get_field_schema():
            logger.warning(f"[WARN] Missing pred_field={pred_field} in {ds_name}, skip.")
            continue
        try:
            rows = evaluate_model_fields(
                dataset=dataset, ds_name=ds_name, subdir_path=subdir_path,
                ckpt_path=ckpt_path, model_tag=model_tag, pred_field=pred_field,
                thresholds=confidence_thresholds, only_matches=False,
            )
            all_rows.extend(rows)
            logger.info(f"[EVAL] 完成: {model_tag} | {ds_name} | {len(rows)} rows")
        except Exception as e:
            logger.error(f"评估失败 model={model_tag}, dataset={ds_name}: {e}")

df = pd.DataFrame(all_rows)
if not df.empty:
    df = df.sort_values(["dataset_name", "model_tag", "confidence_threshold"]).reset_index(drop=True)

logger.info(f"Step 5 完成：共 {len(all_rows)} 条评估记录")
display(df.head(20))
```

# ── 8. 添加 Best F1 / Best Recall 列 ────────────────────

```python
# ── Step 6：计算 best 列（Best F1 / Best Recall@Precision>=gate）──
# 输入：df（Step 5 产出）
# 输出：df 新增 is_best_f1_in_group、is_best_recall_at_prec_gate 列

logger.info("Step 6 开始：计算 best 列")

import numpy as np

if not df.empty:
    group_cols = ["dataset_name", "model_tag"]
    df["precision"] = pd.to_numeric(df["precision"], errors="coerce")
    df["recall"]    = pd.to_numeric(df["recall"],    errors="coerce")

    den = df["precision"] + df["recall"]
    df["f1_calc"] = np.where(den > 0, 2 * df["precision"] * df["recall"] / den, np.nan)

    df["is_best_f1_in_group"] = False
    valid = df[df["f1_calc"].notna()]
    if not valid.empty:
        idx_best = valid.groupby(group_cols)["f1_calc"].idxmax()
        df.loc[idx_best, "is_best_f1_in_group"] = True

    PREC_GATE = 0.90
    df["is_best_recall_at_prec_gate"] = False
    tmp = df[(df["precision"] >= PREC_GATE) & (df["recall"].notna())].copy()
    if not tmp.empty:
        idx_best_recall = tmp.groupby(group_cols)["recall"].idxmax()
        df.loc[idx_best_recall, "is_best_recall_at_prec_gate"] = True
    else:
        logger.warning(f"[WARN] No rows meet precision >= {PREC_GATE}")

logger.info("Step 6 完成：best 列计算结束")
display(df[df.get("is_best_f1_in_group", False)].head(50))
```

```python
# ── Step 7：导出评估结果到 CSV 和 XLSX ─────────────────────
# 输入：df（Step 6 产出）
# 输出：out_csv, out_xlsx

logger.info("Step 7 开始：导出评估结果")

try:
    df.to_csv(out_csv, index=False)
    logger.info(f"已保存 CSV: {out_csv}")

    with pd.ExcelWriter(out_xlsx, engine="openpyxl") as writer:
        df.to_excel(writer, sheet_name="summary", index=False)
        best_f1 = df[df["is_best_f1_in_group"]].copy()
        best_f1.to_excel(writer, sheet_name="best_f1", index=False)
        best_recall = df[df["is_best_recall_at_prec_gate"]].copy()
        best_recall.to_excel(writer, sheet_name="best_recall_at_prec", index=False)
    logger.info(f"已保存 XLSX: {out_xlsx}")
except Exception as e:
    logger.error(f"导出失败: {e}")

logger.info("Step 7 完成：评估结果已导出")
```
