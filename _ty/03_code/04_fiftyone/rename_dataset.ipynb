```python
import fiftyone as fo  

# 批量重命名示例  
dataset_mappings = {  
    "ms1_0726-0809_11_ok_sahi_v1": "v3_sahi_ms1_0726-0809_11_ok",  
    "ms1_0809-0823_34_ok_sahi_v1": "v3_sahi_ms1_0809-0823_34_ok",  
    "ms1_0605-0621_40_ok_sahi_v1": "v3_sahi_ms1_0605-0621_40_ok",
    "ms1_0710-0726_36_ok_sahi_v1": "v3_sahi_ms1_0710-0726_36_ok",
    "ms2_0726-0809_13_ok_sahi_v1": "v3_sahi_ms2_0726-0809_13_ok",
    "ms2_0809-0823_10_ok_sahi_v1": "v3_sahi_ms2_0809-0823_10_ok",
    "sw1_0605-0613_07_ok_sahi_v1": "v3_sahi_sw1_0605-0613_07_ok",
}  

for old_name, new_name in dataset_mappings.items():  
    dataset = fo.load_dataset(old_name)  
    dataset.name = new_name  # 通过name属性重命名  
    print(f"Dataset '{old_name}' renamed to '{new_name}'")
```
