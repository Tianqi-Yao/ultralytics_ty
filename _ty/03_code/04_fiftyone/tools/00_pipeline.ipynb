```python
# list dataset and labels
for dataset_name in fo.list_datasets():
    dataset = fo.load_dataset(dataset_name)
    print(f"Dataset: {dataset.name}")
    print(f"Labels: {dataset.count_sample_tags()}")
    # print(f"summary: {dataset.summary()}")
    sidebar_groups = fo.DatasetAppConfig.default_sidebar_groups(dataset)  
    # print(f"sidebar_groups: {sidebar_groups}")
    for group in sidebar_groups:  
        print(f"分组名称: {group.name}")  
        print(f"字段路径: {group.paths}")  
        print(f"展开状态: {group.expanded}")  
    print("-" * 30)
```
