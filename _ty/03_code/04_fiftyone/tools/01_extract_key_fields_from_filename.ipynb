### 添加focus

```python
import fiftyone as fo
import os


# datasets_list = [
#     'sahi_null_run_whole_rawData_good3_v1_lloyd',
#  'sahi_null_run_whole_rawData_good3_v1_ms1',
#  'sahi_null_run_whole_rawData_good3_v1_southfarm1',
#  'sahi_null_run_whole_rawData_good3_v1_sw1',
#  'sahi_null_run_whole_rawData_good3_v1_sw2',
# ]

for dataset_name in datasets_list:
    try:
        print(f"Processing dataset: {dataset_name}")
        dataset = fo.load_dataset(dataset_name)
        # 为数据集中所有样本添加focus字段
        dataset.add_sample_field("focus", fo.IntField)  # 显式声明字段类型

        # 批量设置值
        focus_values = []
        for sample in dataset.iter_samples(autosave=True):  
            filepath = sample.filepath  
            filename = os.path.basename(filepath)
            # 4656x3496_2025-06-11_09-14-03_focus350.jpg
            focus = int(filename.split("_")[-1].split(".")[0].replace("focus", ""))
            sample["focus"] = focus  
        dataset.create_index("focus")
    except Exception as e:
        print(f"Error processing dataset {dataset_name}: {e}")
```

### 添加日期

```python
# add datetime field
import pandas as pd

for dataset_name in datasets_list:
    try:
        print(f"Processing dataset: {dataset_name}")
        dataset = fo.load_dataset(dataset_name)
        dataset.add_sample_field("datetime", fo.DateTimeField)  # 显式声明字段类型

        for sample in dataset.iter_samples(autosave=True):
            filepath = sample.filepath
            filename = os.path.basename(filepath)
            # 4656x3496_2025-06-11_09-14-03_focus350.jpg
            datetime_str = filename.split("_")[1] + "_" + filename.split("_")[2]
            datetime_obj = pd.to_datetime(datetime_str, format="%Y-%m-%d_%H-%M-%S")
            sample["datetime"] = datetime_obj
        dataset.create_index("datetime")
    except Exception as e:
        print(f"Error processing dataset {dataset_name}: {e}")
```

```python
import fiftyone as fo

# datasets_list = ["air2_0701_0823", "southfarm2_0712_0823"]

for dataset_name in datasets_list:
    dataset = fo.load_dataset(dataset_name)
    sample = next(dataset.iter_samples())

    print(f"{dataset_name}:")

    # 查找所有包含Detection的字段
    label_fields = []
    for field_name in dir(sample):
        if not field_name.startswith('_'):
            field_value = getattr(sample, field_name, None)
            if field_value and hasattr(field_value, 'detections'):
                label_fields.append(field_name)

    # 显示所有标签字段
    if label_fields:
        for field_name in label_fields:
            field_value = getattr(sample, field_name)
            if field_value.detections:  # 确保有检测
                first_label = field_value.detections[0].label
                print(f"    \"{field_name}\",")
    else:
        print("  未找到标签字段")

    print()
```

#### 添加预测数量-- ALL

```python
import os
import fiftyone as fo

labels_list = [
    "pred_yolo11l_20pct_null_images_add_rawData_batch_4_final",
]

for dataset_name in datasets_list:
    dataset = fo.load_dataset(dataset_name)

    for pred_field in labels_list:
        count_field = f"{pred_field}__count"   # 新建一个数量字段（避免覆盖原预测字段）

        # 1) 声明 int 字段（避免重复 add 报错）
        if count_field not in dataset.get_field_schema():
            dataset.add_sample_field(count_field, fo.IntField)

        # 2) 逐样本写入预测数量
        for sample in dataset.iter_samples(autosave=True):
            preds = getattr(sample, pred_field, None)   # 预测结果（Detections）

            if preds is None or preds.detections is None:
                detections_len = 0
            else:
                detections_len = len(preds.detections)

            sample[count_field] = detections_len

        # 3) 建索引（可选）
        dataset.create_index(count_field)
```

#### 添加预测数量 -- 0.5


```python
import fiftyone as fo

labels_list = [
    "pred_yolo11l_20pct_null_images_add_rawData_batch_4_final",
]

for dataset_name in datasets_list:
    dataset = fo.load_dataset(dataset_name)

    for pred_field in labels_list:
        count_field = f"{pred_field}__count_0.5"

        # 1) 声明 int 字段（避免重复 add 报错）
        if count_field not in dataset.get_field_schema():
            dataset.add_sample_field(count_field, fo.IntField)

        # 2) 逐样本写入预测数量
        for sample in dataset.iter_samples(autosave=True):
            preds = getattr(sample, pred_field, None)

            if preds is None or preds.detections is None:
                detections_len = 0
            else:
                # 只统计 confidence >= 0.5 的检测框
                detections_len = sum(
                    1 for d in preds.detections
                    if d.confidence is not None and d.confidence >= 0.5
                )

            sample[count_field] = detections_len

        # 3) 建索引（可选）
        dataset.create_index(count_field)
```
