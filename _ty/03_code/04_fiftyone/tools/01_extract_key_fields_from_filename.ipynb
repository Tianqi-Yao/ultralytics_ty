### 添加focus

```python
import fiftyone as fo
import os


# datasets_list = [
#     'sahi_null_run_whole_rawData_good3_v1_lloyd',
#  'sahi_null_run_whole_rawData_good3_v1_ms1',
#  'sahi_null_run_whole_rawData_good3_v1_southfarm1',
#  'sahi_null_run_whole_rawData_good3_v1_sw1',
#  'sahi_null_run_whole_rawData_good3_v1_sw2',
# ]

for dataset_name in datasets_list:
    try:
        print(f"Processing dataset: {dataset_name}")
        dataset = fo.load_dataset(dataset_name)
        # 为数据集中所有样本添加focus字段
        dataset.add_sample_field("focus", fo.IntField)  # 显式声明字段类型

        # 批量设置值
        focus_values = []
        for sample in dataset.iter_samples(autosave=True):  
            filepath = sample.filepath  
            filename = os.path.basename(filepath)
            # 4656x3496_2025-06-11_09-14-03_focus350.jpg
            focus = int(filename.split("_")[-1].split(".")[0].replace("focus", ""))
            sample["focus"] = focus  
        dataset.create_index("focus")
    except Exception as e:
        print(f"Error processing dataset {dataset_name}: {e}")
```

### 添加日期

```python
# add datetime field
import pandas as pd

for dataset_name in datasets_list:
    try:
        print(f"Processing dataset: {dataset_name}")
        dataset = fo.load_dataset(dataset_name)
        dataset.add_sample_field("datetime", fo.DateTimeField)  # 显式声明字段类型

        for sample in dataset.iter_samples(autosave=True):
            filepath = sample.filepath
            filename = os.path.basename(filepath)
            # 4656x3496_2025-06-11_09-14-03_focus350.jpg
            datetime_str = filename.split("_")[1] + "_" + filename.split("_")[2]
            datetime_obj = pd.to_datetime(datetime_str, format="%Y-%m-%d_%H-%M-%S")
            sample["datetime"] = datetime_obj
        dataset.create_index("datetime")
    except Exception as e:
        print(f"Error processing dataset {dataset_name}: {e}")
```

```python
import fiftyone as fo
from pathlib import Path
# /media/tianqi/16tb/SWD/01_Data/2025_SWD_data_RAW/01_North/air1/air1_0427-0611/output/ManualFocus_air1_0427-0611/4656x3496_2025-06-07_09-03-38_focus675.jpg

# 添加field_segment字段，例如ManualFocus_air1_0427-0611则为 air1_0427-0611
# 再添加field字段，例如ManualFocus_air1_0427-0611则为 air1
for dataset_name in datasets_list:
    try:
        print(f"Processing dataset: {dataset_name}")
        dataset = fo.load_dataset(dataset_name)
        dataset.add_sample_field("field_segment", fo.StringField)  # 显式声明字段类型
        dataset.add_sample_field("field", fo.StringField)  # 显式声明字段类型

        for sample in dataset.iter_samples(autosave=True):
            filepath = Path(sample.filepath)
            filename = filepath.name
            filefoldername = filepath.parent.name
            field_segment_value = filefoldername.split("_", 1)[1]  # ManualFocus_air1_0427-0611 -> air1_0427-0611
            field_value = field_segment_value.split("_")[0]  # air1_0427-0611 -> air1
            sample["field_segment"] = field_segment_value
            sample["field"] = field_value
        dataset.create_index("field_segment")
        dataset.create_index("field")
    except Exception as e:
        print(f"Error processing dataset {dataset_name}: {e}")
```

#### 添加预测数量-- ALL

```python
import os
import fiftyone as fo

labels_list = [
    "pred_yolo11l_20pct_null_images_add_rawData_batch_4_final",
]

for dataset_name in datasets_list:
    dataset = fo.load_dataset(dataset_name)

    for pred_field in labels_list:
        count_field = f"{pred_field}__count"   # 新建一个数量字段（避免覆盖原预测字段）

        # 1) 声明 int 字段（避免重复 add 报错）
        if count_field not in dataset.get_field_schema():
            dataset.add_sample_field(count_field, fo.IntField)

        # 2) 逐样本写入预测数量
        for sample in dataset.iter_samples(autosave=True):
            preds = getattr(sample, pred_field, None)   # 预测结果（Detections）

            if preds is None or preds.detections is None:
                detections_len = 0
            else:
                detections_len = len(preds.detections)

            sample[count_field] = detections_len

        # 3) 建索引（可选）
        dataset.create_index(count_field)
```

#### 添加预测数量 -- 0.5


```python
import fiftyone as fo

labels_list = [
    # "pred_yolo11l_20pct_null_images_add_rawData_batch_4_final",
    "pred_yolo11m_20pct_null_images_add_rawData_batch_16_final",
]

datasets_list = [
    "2025_north_v1__air1",
    "2025_north_v1__air2",
    "2025_north_v1__jeff",
    "2025_north_v1__lloyd1",
    "2025_north_v1__lloyd2",
    "2025_north_v1__southfarm1",
    "2025_north_v1__southfarm2",
]

for dataset_name in datasets_list:
    print(f"Processing dataset: {dataset_name}")
    dataset = fo.load_dataset(dataset_name)

    for pred_field in labels_list:
        count_field = f"{pred_field}__count_pct50"

        # 1) 声明 int 字段（避免重复 add 报错）
        if count_field not in dataset.get_field_schema():
            dataset.add_sample_field(count_field, fo.IntField)

        # 2) 逐样本写入预测数量
        for sample in dataset.iter_samples(autosave=True):
            preds = getattr(sample, pred_field, None)

            if preds is None or preds.detections is None:
                detections_len = 0
            else:
                # 只统计 confidence >= 0.5 的检测框
                detections_len = sum(
                    1 for d in preds.detections
                    if d.confidence is not None and d.confidence >= 0.5
                )

            sample[count_field] = detections_len

        # 3) 建索引（可选）
        dataset.create_index(count_field)
```

# 创建新的dataset

```python
new_datasets_list = [
    "2025_north_v1__air1",
    "2025_north_v1__air2",
    "2025_north_v1__jeff",
    "2025_north_v1__lloyd1",
    "2025_north_v1__lloyd2",
    "2025_north_v1__southfarm1",
    "2025_north_v1__southfarm2",
]
```

```python
import fiftyone as fo


new_dataset_counter = 0
total_dataset_counter = 0
for new_dataset_name in new_datasets_list:
    print(f"Dataset name: {new_dataset_name}")
    # 创建新的空数据集  
    new_dataset = fo.Dataset(new_dataset_name)
    dataset_counter = 0
    for dataset_name in datasets_list:
        print(f"  Merging dataset: {dataset_name}")
        dataset = fo.load_dataset(dataset_name)
        if "air1" in dataset_name and "air1" in new_dataset_name:
            new_dataset.merge_samples(dataset)
        elif "air2" in dataset_name and "air2" in new_dataset_name:
            new_dataset.merge_samples(dataset)
        elif "jeff" in dataset_name and "jeff" in new_dataset_name:
            new_dataset.merge_samples(dataset)
        elif "lloyd1" in dataset_name and "lloyd1" in  new_dataset_name:
            new_dataset.merge_samples(dataset)
        elif "lloyd2" in dataset_name and "lloyd2" in  new_dataset_name:
            new_dataset.merge_samples(dataset)
        elif "southfarm1" in dataset_name and "southfarm1" in  new_dataset_name:
            new_dataset.merge_samples(dataset)
        elif "southfarm2" in dataset_name and "southfarm2" in  new_dataset_name:
            new_dataset.merge_samples(dataset)
        else:
            print(f"    Skipping dataset {dataset_name} as it does not match the new dataset name {new_dataset_name}.")
            continue
        dataset_counter += 1
        total_dataset_counter += 1
        print(f"    Merged {dataset_counter} datasets into {new_dataset_name} so far.")
    new_dataset_counter += 1
    print(f"  Finished merging for {new_dataset_name}. Total merged datasets so far: {new_dataset_counter}\n")
print(f"Finished merging all datasets. Total new datasets created: {new_dataset_counter}. Total datasets processed: {total_dataset_counter}.")
```
