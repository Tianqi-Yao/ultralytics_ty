### 添加focus

```python
import fiftyone as fo
import os


datasets_list = [
    'test_run2_image',
]

for dataset_name in datasets_list:
    dataset = fo.load_dataset(dataset_name)
    # 为数据集中所有样本添加focus字段
    dataset.add_sample_field("focus", fo.IntField)  # 显式声明字段类型

    # 批量设置值
    focus_values = []
    for sample in dataset.iter_samples(autosave=True):  
        filepath = sample.filepath  
        filename = os.path.basename(filepath)  
        focus = int(filename.split("_")[-1].split(".")[0])  
        sample["focus"] = focus  
    dataset.create_index("focus")  
```

### 添加日期

```python
# add datetime field
import pandas as pd

for dataset_name in datasets_list:
    dataset = fo.load_dataset(dataset_name)
    dataset.add_sample_field("datetime", fo.DateTimeField)  # 显式声明字段类型

    for sample in dataset.iter_samples(autosave=True):
        filepath = sample.filepath
        filename = os.path.basename(filepath)
        datetime_str = "2024" + filename.split("_")[0] + "_" + filename.split("_")[1]
        datetime_obj = pd.to_datetime(datetime_str, format="%Y%m%d_%H%M")
        sample["datetime"] = datetime_obj
    dataset.create_index("datetime")
```

```python
import os
import fiftyone as fo

labels_list = [
    "pred_yolo11l_20pct_null_images_add_rawData_batch_16_final",
    "pred_yolo11l_20pct_null_images_add_rawData_batch_4_final",
    "pred_yolo11l_20pct_null_images_add_rawData_batch_8_final",
    "pred_yolo11m_20pct_null_images_add_rawData_batch_16_final",
    "pred_yolo11m_20pct_null_images_add_rawData_batch_4_final",
    "pred_yolo11m_20pct_null_images_add_rawData_batch_8_final",
    "pred_yolo11n_20pct_null_images_add_rawData_batch_16_final",
    "pred_yolo11n_20pct_null_images_add_rawData_batch_4_final",
    "pred_yolo11n_20pct_null_images_add_rawData_batch_8_final",
    "pred_yolo11s_20pct_null_images_add_rawData_batch_16_final",
    "pred_yolo11s_20pct_null_images_add_rawData_batch_4_final",
    "pred_yolo11s_20pct_null_images_add_rawData_batch_8_final",
    "pred_yolo11x_20pct_null_images_add_rawData_batch_4_final",
    "pred_yolo11x_20pct_null_images_add_rawData_batch_8_final",

]

for dataset_name in datasets_list:
    dataset = fo.load_dataset(dataset_name)

    for pred_field in labels_list:
        count_field = f"{pred_field}__count"   # 新建一个数量字段（避免覆盖原预测字段）

        # 1) 声明 int 字段（避免重复 add 报错）
        if count_field not in dataset.get_field_schema():
            dataset.add_sample_field(count_field, fo.IntField)

        # 2) 逐样本写入预测数量
        for sample in dataset.iter_samples(autosave=True):
            preds = getattr(sample, pred_field, None)   # 预测结果（Detections）

            if preds is None or preds.detections is None:
                detections_len = 0
            else:
                detections_len = len(preds.detections)

            sample[count_field] = detections_len

        # 3) 建索引（可选）
        dataset.create_index(count_field)
```
