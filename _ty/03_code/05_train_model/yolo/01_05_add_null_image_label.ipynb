### 给 null images 批量生成空 label 文件

```python
from pathlib import Path

def ensure_empty_labels(images_dir: Path, labels_dir: Path, exts=(".jpg",".jpeg",".png",".bmp",".tif",".tiff",".webp")):
    images_dir = Path(images_dir)
    labels_dir = Path(labels_dir)
    labels_dir.mkdir(parents=True, exist_ok=True)

    imgs = [p for p in images_dir.rglob("*") if p.suffix.lower() in exts]
    created = 0

    for img in imgs:
        # label 路径：同名 txt
        label_path = labels_dir / (img.stem + ".txt")
        if not label_path.exists():
            label_path.write_text("")  # 空文件
            created += 1

    print(f"Checked {len(imgs)} images, created {created} empty label files in {labels_dir}")

# 示例：按 split 分别跑
ensure_empty_labels(Path("dataset/images/train"), Path("dataset/labels/train"))
ensure_empty_labels(Path("dataset/images/val"),   Path("dataset/labels/val"))
```
