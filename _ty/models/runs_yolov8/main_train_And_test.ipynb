```python
import time
import subprocess

GPU_ID = 0

while True:
    try:
        q = subprocess.check_output(
            ["nvidia-smi", f"--id={GPU_ID}",
             "--query-gpu=memory.used",
             "--format=csv,noheader,nounits"],
            stderr=subprocess.STDOUT, text=True
        ).strip()
        used_mb = float(q)
        if used_mb < 2000:  # å°äºŽ 2GB è®¤ä¸ºæ˜¯ç©ºé—²
            print(f"[ok] GPU{GPU_ID} ç©ºé—²ï¼Œå¼€å§‹è®­ç»ƒï¼")
            break
        else:
            print(f"[wait] GPU{GPU_ID} å ç”¨ {used_mb:.0f}MBï¼Œç­‰å¾…ä¸­...")
    except Exception as e:
        print(f"[check] èŽ·å– GPU çŠ¶æ€å¤±è´¥: {e}")

    time.sleep(600)  # æ¯éš” 60s å†æŸ¥ä¸€æ¬¡
```

```python
from ultralytics import YOLO


# ==== ä½ éœ€è¦æ‰‹åŠ¨å¡«å†™çš„åˆ—è¡¨ ====
yamlFileNames = [  # æ•°æ®é›† yaml æ–‡ä»¶åï¼ˆä¸è¦å†™åŽç¼€ .yamlï¼‰
    # "data_split_0.4_0.3_0.3",
    # "data_split_0.5_0.2_0.3",
    "data_split_0.5_0.3_0.2",
    "data_split_0.6_0.1_0.3",
    "data_split_0.6_0.2_0.2",
]
batchSizes = [
    4, 8, 16, 32, 64,
]
models = [
    "yolov8n-seg.pt",
    "yolov8s-seg.pt",
    # "yolov8m-seg.pt",
    "yolov8n.pt",
    "yolov8s.pt",
    # "yolov8m-seg.pt",
    # "yolo8l-seg.pt", "yolo8x-seg.pt",
    # "yolo8l.pt", "yolo8x.pt",
]
# =================================

runPath = "/workspace/models/runs_yolov8/yaml/"

# å¾ªçŽ¯éåŽ†
for yamlFileName in yamlFileNames:
    yamlPath = runPath + yamlFileName + ".yaml"
    for modelFile in models:
        for batch in batchSizes:
            print(f"\nðŸš€ Training model={modelFile}, dataset={yamlFileName}, batch={batch}")

            model = YOLO(modelFile)

            model.train(
                data=yamlPath,
                epochs=1000,
                imgsz=640,
                batch=batch,
                device=0,
                workers=4,
                project=f"output/{modelFile}",   # è¾“å‡ºç›®å½•
                name=f"{yamlFileName}_{batch}",  # run åå­—æ›´æ¸…æ™°
            )

            # æµ‹è¯•é›†éªŒè¯
            model.val(
                data=yamlPath,
                split="test",    # é€‰ä¸­testæ•°æ®é›†
                name=f"{yamlFileName}_{batch}_test"
            )
```
