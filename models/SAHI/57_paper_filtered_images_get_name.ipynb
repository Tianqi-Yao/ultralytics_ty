```python
# read csv
import pandas as pd
df = pd.read_csv("r2.csv")
df.head()
```

```python
filename_df = df['filename'].tolist()  # Note the trailing space in the column name
print("filename:", filename_df, len(filename_df))
```

```python
# 拆分成jeff_0613-0624_04_ok_0614_0602_580_track_vis.jpg 文件夹名和文件名 jeff_0613-0624_04_ok 0614_0602_580_track_vis.jpg
import os

# folder, file = filename.rsplit("_", 4)[0], "_".join(filename.rsplit("_", 4)[1:])
folders = [os.path.splitext(f)[0].rsplit("_", 5)[0] for f in filename_df]
files = ["_".join(os.path.splitext(f)[0].rsplit("_", 5)[1:]) + os.path.splitext(f)[1] for f in filename_df]
display("folders:", folders, len(folders))
display("files:", files, len(files))
```

```python
# 遍历文件夹， 举例：原本/workspace/models/SAHI/run_v7/jeff_0613-0624_04_ok/raw_data_sliced_merge/pose_and_det_v1/10_visualization_tracking_results/vis/0614_0602_580_track_vis.jpg找到对应的文件/workspace/models/SAHI/run_v7/jeff_0613-0624_04_ok/raw_data/0614_0602_580.jpg
from itertools import count
import os
import shutil

base_path = "/workspace/models/SAHI/run_v7"
output_path = "/workspace/models/SAHI/filtered_images"
os.makedirs(output_path, exist_ok=True)

count = 0
for folder, file in zip(folders, files):
    # 原始文件名：去掉 "_track_vis" 得到对应的 raw_data 文件名
    raw_name = file.replace("_track_vis", "")
    
    # 构造路径
    src = os.path.join(base_path, folder, "raw_data", raw_name)
    dst = os.path.join(output_path, raw_name)
    
    # 如果存在就复制
    if os.path.exists(src):
        # shutil.copy(src, dst)
        print(f"Copied: {src} -> {dst}")
        count += 1
    else:
        print(f"Not found: {src}")

print(f"Total copied files: {count}")
```
